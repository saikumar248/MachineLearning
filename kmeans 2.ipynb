```python
from math import *
import matplotlib.pyplot as plt
c1=(1.678,1.712)
c2=(0.809,1.349)
c3=(1.112,1.679)
c1_l=[(1.678,1.712)]
c2_l=[(0.809,1.349)]
c3_l=[(1.112,1.679)]
def ans(x):
    global c1,c2,c3
    c1_dist=sqrt((c1[0]-x[0])**2+(c1[1]-x[1])**2)
    
    c2_dist=sqrt((c2[0]-x[0])**2+(c2[1]-x[1])**2)
    
    c3_dist=sqrt((c3[0]-x[0])**2+(c3[1]-x[1])**2)
    
    minimum=min(c1_dist,c2_dist,c3_dist)
    if minimum == c1_dist:
        print("Belongs to cluster 1")
    
    if minimum == c2_dist:
        print("Belongs to cluster 2")
    
    if minimum == c3_dist:
        print("Belongs to cluster 3")

def distance(x):
    global c1,c2,c3,c1_l,c2_l,c3_l
    c1_dist=sqrt((c1[0]-x[0])**2+(c1[1]-x[1])**2)
    
    c2_dist=sqrt((c2[0]-x[0])**2+(c2[1]-x[1])**2)
    
    c3_dist=sqrt((c3[0]-x[0])**2+(c3[1]-x[1])**2)
    
    minimum=min(c1_dist,c2_dist,c3_dist)
    if minimum == c1_dist:
        c1=((c1[0]+x[0])/2,(c1[1]+x[1])/2)
        c1_l.append(x)
    
    if minimum == c2_dist:
        c2=((c2[0]+x[0])/2,(c2[1]+x[1])/2)
        c2_l.append(x)
    
    if minimum == c3_dist:
        c3=((c3[0]+x[0])/2,(c3[1]+x[1])/2)
        c3_l.append(x)

    return c1,c2,c3
 
# print(c1,c2,c3)

datapoints=[(1.713,1.586),(0.18,1.786),(0.353,1.24),(0.95,1.566),(1.486,0.759),(1.266,1.106),(1.54,0.419),(0.459,1.799),(0.773,0.186)]
find_point =(0.906,0.606)
for i in datapoints:
    distance(i)
# print(lst)
# print(c1,c2,c3)
ans(find_point)

x1,y1,x2,y2,x3,y3=[],[],[],[],[],[]
for i in c1_l:
    x1.append(i[0])
    y1.append(i[1])
for i in c2_l:
    x2.append(i[0])
    y2.append(i[1])
for i in c3_l:
    x3.append(i[0])
    y3.append(i[1])
plt.scatter(x1,y1,s=25,label="cluster1")
plt.scatter(x2,y2,s=25,label="cluster2")
plt.scatter(x3,y3,s=25,label="cluster3")
plt.scatter([0.906],[0.606],s=40,label="findpoint")
plt.legend()
plt.show()
```
