```python
# import matplotlib.pyplot as plt
import numpy as np


from math import *
dist_0 =[]
dist_1 =[]
def calculate_0(x,p):
    for i in x:
        dist=sqrt((p[0]-i[0])**2+(p[1]-i[1])**2)
        dist_0.append(dist)
def calculate_1(x,p):
    for i in x:
        dist=sqrt((p[0]-i[0])**2+(p[1]-i[1])**2)
        dist_1.append(dist)
a=int(input("Enter 1st Datapoint"))
b=int(input("Enter 2nd Datapoint"))
p=(a,b)
points_0=[(2,1),(2,1),(1,0),(2,2)]
points_1=[(0,1),(1,2),(1,0),(0,0),(2,0),(1,2),(0,2)]
calculate_0(points_0,p)
calculate_1(points_1,p)
print(dist_0)
print(dist_1)


plt.plot(dist_0)
print(plt.show())

final_list=dist_0+dist_1
sort_dist=sorted(final_list)
print(sort_dist)
count=0
count1=0
for i in range(len(sort_dist)):
    if sort_dist[i] in dist_0:
        count+=1
    if sort_dist[i] in dist_1:
        count1+=1
    
if(count>count1):
    print("Belongs to 0 Label")
else:
    print("Belongs to 1 Label")

    
```
